```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data
# Create a dataset with 3 clusters using `make_blobs`.
# Parameters:
# - n_samples: Total number of data points
# - n_features: Number of features (dimensions) of the data
# - centers: Number of cluster centers to generate
# - random_state: Seed for reproducibility
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42

X, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# Visualize the raw data before clustering
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='gray', s=50, label='Raw Data')
plt.title("Raw Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# Step 2: Initialize the K-Means model
# Parameters:
# - n_clusters: Number of clusters to form
# - random_state: Seed for reproducibility
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)

# Step 3: Fit the K-Means model to the data
# The `fit` method computes the centroids of the clusters.
kmeans.fit(X)

# Step 4: Extract results
# Centroids of the clusters (calculated by K-Means)
centroids = kmeans.cluster_centers_

# Cluster labels assigned to each data point
# - Each data point is assigned to the nearest cluster center
y_kmeans = kmeans.labels_

# Step 5: Visualize the clustered data
plt.figure(figsize=(8, 6))

# Scatter plot of data points, colored by their assigned cluster
# `y_kmeans` determines the cluster assignment of each point
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', marker='o', s=50, label='Data points')

# Scatter plot of centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, label='Centroids', edgecolor='black')

# Add titles and labels
plt.title("K-Means Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()

# Step 6: Evaluate clustering performance
# Calculate the Inertia: The sum of squared distances of samples to their nearest cluster center
# Lower inertia means that the clusters are tighter and more cohesive.
inertia = kmeans.inertia_
print(f"Inertia (Sum of Squared Distances to Centroids): {inertia:.2f}")

# Optional: Elbow Method to Determine Optimal Number of Clusters
# The Elbow Method involves running K-Means with varying `n_clusters` and plotting the inertia
inertia_values = []
cluster_range = range(1, 10)  # Test for 1 to 9 clusters

for n in cluster_range:
    kmeans_temp = KMeans(n_clusters=n, random_state=random_state)
    kmeans_temp.fit(X)
    inertia_values.append(kmeans_temp.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia_values, marker='o')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.xticks(cluster_range)
plt.show()
```
